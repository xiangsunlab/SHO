```python
#Mass check
nbeads = 16
mass = 2
stgmass = np.loadtxt('stgmass')
print(stgmass[0])
for i in (np.arange(1,16,1)):
    print(mass * (((i + 1) / i)) - stgmass[i][0], 1 / (mass * (((i + 1) / i))) - stgmass[i][1])
```

```python
disthund = np.loadtxt('disthund')
x = np.arange(-50,50,0.01)
p = np.arange(-50,50,0.01)
xsim = disthund[:,0]
psim = disthund[:,1]
```

```python
w = 1
mass = 2
hbar = 1
beta = 1
bhw = beta * hbar * w
mw = mass * w
x = np.arange(-50,50,0.01)

rho = sqrt(mw/(2*np.pi*hbar*sinh(bhw))) * exp(- (mw/(2*hbar*sinh(bhw)))*(2*x**2 * cosh(bhw) -2*x**2))
plt.plot(x,rho/sum(rho)/0.01,label = "Theoretical distribution 100 traj")
plt.plot(x,xsim/sum(xsim)/0.01)
print(sum(rho))
plt.xlim(-5,5)
plt.legend()
```
