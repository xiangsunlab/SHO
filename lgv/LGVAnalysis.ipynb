```python
config = np.loadtxt('lgv_config')
config
```

```python
x = np.arange(-50,50,0.01)
p = np.arange(-50,50,0.01)
beta = 1
m = 2
omega = 1
Zx = sum(np.exp(- beta * x * x * m * omega * omega * 0.5))
print(Zx)
CDFx_theory = np.exp(- beta * x * x * m * omega * omega/2) / Zx / 0.01
print(sum(CDFx_theory))
Zp = sum(np.exp(- beta * p * p / m * 0.5))
print(Zp)
CDFp_theory = np.exp(- beta * p * p / m * 0.5) / Zp / 0.01
print(sum(CDFp_theory))
```

```python
sum(config[:,0]),sum(config[:,1])
```

```python
CDF_x = config[:,0] / sum(config[:,0]) /0.01
CDF_p = config[:,1] / sum(config[:,1]) /0.01
```

```python
plt.plot(x,CDF_x,label = 'NVT simulation: Position')
plt.plot(x,CDFx_theory,label = 'NVT theoretical position distribution')
plt.xlim(-5,5)
plt.plot()
plt.legend()
```

```python
plt.plot(p,CDF_p,label = 'NVT simulation: momenta')
plt.plot(p,CDFp_theory,label = 'NVT theoretical momenta distribution')
plt.xlim(-5,5)
plt.plot()
plt.legend()
```
